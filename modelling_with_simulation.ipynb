### SIMULATING MONTHLY MOVEMENT

```python
import pandas as pd
import numpy as np

def split_population_and_age_groups(row):
    new_rows = []
    total_population = row['population']
    age_columns = ['0-4-f', '0-4-m', '5-11-f', '5-11-m', '12-17-f', '12-17-m', '18-59-f', '18-59-m', '60+-f', '60+-m']

    # Generate random splits for the total population
    pop_splits = np.random.rand(12)
    pop_splits /= pop_splits.sum()
    monthly_populations = (pop_splits * total_population).astype(int)

    # Ensure the sum of monthly populations matches the original total population
    diff = total_population - monthly_populations.sum()
    # Distribute the difference randomly
    indices_to_adjust = np.random.choice(12, size=abs(diff), replace=True)
    monthly_populations[indices_to_adjust] += np.sign(diff)

    for i in range(12):
        new_row = row.copy()
        new_row['duration'] = f'Month {i + 1}'
        new_row['population'] = monthly_populations[i]

        # Split the age group populations based on the split ratio of the total population
        # This assumes that the proportion of each age group remains roughly constant across months
        # Alternatively, one could generate independent splits for each age group and ensure they sum up
        # but this approach simplifies the constraint where total population must equal sum of age groups.
        age_group_monthly_sum = 0
        for col in age_columns:
            original_age_pop = row[col]
            # Split the age group population using the same split ratio as the total population
            monthly_age_pop = int(original_age_pop * pop_splits[i])
            new_row[col] = monthly_age_pop
            age_group_monthly_sum += monthly_age_pop

        # Adjust age group populations to exactly match the monthly total population
        diff_age_sum = monthly_populations[i] - age_group_monthly_sum
        if diff_age_sum != 0:
            # Distribute the difference randomly among age groups
            age_cols_to_adjust = np.random.choice(age_columns, size=abs(diff_age_sum), replace=True)
            for col in age_cols_to_adjust:
                new_row[col] += np.sign(diff_age_sum)

        new_rows.append(new_row)
    return new_rows

# Apply the function to each row
synthetic_data_list = movement_modified.apply(split_population_and_age_groups, axis=1).tolist()

# Flatten the list of lists
synthetic_data_list_flat = [item for sublist in synthetic_data_list for item in sublist]

# Create the new DataFrame
synthetic_movement_df = pd.DataFrame(synthetic_data_list_flat)

print(synthetic_movement_df.head())
print(synthetic_movement_df.info())

# Verification: Check if total population matches the sum of age groups for a few random rows
sample_rows = synthetic_movement_df.sample(5)
age_columns = ['0-4-f', '0-4-m', '5-11-f', '5-11-m', '12-17-f', '12-17-m', '18-59-f', '18-59-m', '60+-f', '60+-m']
sample_rows['age_group_sum'] = sample_rows[age_columns].sum(axis=1)

print("\nVerification Sample:")
print(sample_rows[['population', 'age_group_sum']])
assert (sample_rows['population'] == sample_rows['age_group_sum']).all()
print("\nVerification successful: population matches sum of age groups for sample rows.")
```

```python
import pandas as pd
import reverse_geocoder as rg
import pycountry
import sys

# Check for required libraries
try:
    import pandas
    import reverse_geocoder
    import pycountry
except ImportError:
    print("Error: Required libraries (pandas, reverse_geocoder, pycountry) are not installed.")
    print("Please install them in Grok Studio (e.g., 'pip install pandas pycountry reverse_geocoder').")
    sys.exit(1)

# Load the dataset
try:
    df = locations.copy()
except FileNotFoundError:
    print("Error: 'locations.csv' not found. Ensure the file is in the working directory.")
    sys.exit(1)

# Verify required columns
if not {'latitude', 'longitude'}.issubset(df.columns):
    print("Error: Dataset must contain 'latitude' and 'longitude' columns.")
    sys.exit(1)

# Function to get country info from coordinates
def get_country_info(lat, lon):
    try:
        # Perform reverse geocoding
        results = rg.search([(lat, lon)], mode=1)  # mode=1 for faster single query
        country_code = results[0]['cc']  # Get 2-letter country code
        # Convert to country name and ISO3 code
        country = pycountry.countries.get(alpha_2=country_code)
        if country:
            return pd.Series({
                'country_name': country.name,
                'iso3': country.alpha_3
            })
        else:
            return pd.Series({
                'country_name': None,
                'iso3': None
            })
    except Exception as e:
        print(f"Warning: Failed to geocode ({lat}, {lon}): {e}")
        return pd.Series({
            'country_name': None,
            'iso3': None
        })

# Apply geocoding to get country name and ISO3 code
print("Performing reverse geocoding...")
df[['country_name', 'iso3']] = df.apply(
    lambda row: get_country_info(row['latitude'], row['longitude']),
    axis=1
)

# Drop rows with missing country info
df = df.dropna(subset=['country_name', 'iso3'])

# Keep only the first row per country (based on iso3)
df_unique = df.drop_duplicates(subset=['iso3'], keep='first')

# Select relevant columns
df_unique = df_unique[['id', 'latitude', 'longitude', 'elevation', 'country_name', 'iso3']]

# Save the filtered dataset
output_file = 'locations_unique_countries.csv'
df_unique.to_csv(output_file, index=False)

# Print summary
print(f"Processed {len(df)} locations.")
print(f"Found {len(df_unique)} unique countries.")
print(f"Saved filtered dataset to '{output_file}'.")
print("\nFirst few rows of the result:")
print(df_unique.head())
```

```python
import pandas as pd
import sys
from datetime import datetime
from meteostat import Monthly
import warnings

# Get unique ISO3 codes from movement_with_conflict
valid_iso3 = movement_with_conflict['origin_location_code'].unique()
print(f"Found {len(valid_iso3)} unique origin_location_code values.")

# Load the unique countries dataset
try:
    df_stations = pd.read_csv('locations_unique_countries.csv')
except FileNotFoundError:
    print("Error: 'locations_unique_countries.csv' not found. Ensure the file is in the working directory.")
    sys.exit(1)

# Verify required columns
required_cols = {'id', 'country_name', 'iso3'}
if not required_cols.issubset(df_stations.columns):
    print(f"Error: Dataset must contain {required_cols} columns.")
    sys.exit(1)

# Filter stations to include only those with iso3 in valid_iso3
df_stations = df_stations[df_stations['iso3'].isin(valid_iso3)]
if df_stations.empty:
    print("Error: No stations match the ISO3 codes in movement_with_conflict['origin_location_code'].")
    print("Check if origin_location_code contains valid ISO3 codes matching locations_unique_countries.csv.")
    sys.exit(1)
print(f"Filtered to {len(df_stations)} stations matching origin_location_code.")

# Set time period
start = datetime(2015, 1, 1)
end = datetime.now()  # Fetch up to the latest available date

# Function to fetch monthly data for a station
def fetch_monthly_data(station_id):
    try:
        # Fetch monthly data
        data = Monthly(station_id, start, end)
        data = data.fetch()
        if data.empty:
            print(f"Warning: No monthly data available for station {station_id}")
            return None
        # Reset index to make time a column
        data = data.reset_index()
        # Add station ID
        data['id'] = station_id
        return data
    except Exception as e:
        print(f"Warning: Failed to fetch data for station {station_id}: {e}")
        return None

# Initialize empty list to store data
all_data = []

# Fetch data for each station
print("Fetching monthly weather data...")
for idx, row in df_stations.iterrows():
    station_id = row['id']
    print(f"Processing station {station_id} ({row['country_name']})")
    monthly_data = fetch_monthly_data(station_id)
    if monthly_data is not None:
        # Add country info
        monthly_data['iso3'] = row['iso3']
        monthly_data['country_name'] = row['country_name']
        all_data.append(monthly_data)

# Combine all data
if not all_data:
    print("Error: No data retrieved for any station. Check station IDs or data availability.")
    sys.exit(1)

df_weather = pd.concat(all_data, ignore_index=True)

# Rename and select columns
df_weather = df_weather.rename(columns={'time': 'date'})
# Extract year and month
df_weather['year'] = df_weather['date'].dt.year
df_weather['month'] = df_weather['date'].dt.month
# Select desired columns
columns = ['id', 'iso3', 'country_name', 'year', 'month', 'tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
df_weather = df_weather[columns]

# Drop rows where all weather metrics are NaN'
weather_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
df_weather = df_weather.dropna(subset=weather_cols, how='all')

# Save the dataset
output_file = 'monthly_weather_filtered_by_movement.csv'
df_weather.to_csv(output_file, index=False)

# Print summary
print(f"\nProcessed {len(df_stations)} stations.")
print(f"Retrieved data for {len(df_weather['iso3'].unique())} unique countries.")
print(f"Saved dataset to '{output_file}' with {len(df_weather)} rows.")
print("\nFirst few rows of the result:")
print(df_weather.head())
```

```python
import pandas as pd
import numpy as np
def simulate_resources(df):
    """
    Simulates available and added resources based on population and asylum country.

    Args:
        df (pd.DataFrame): The input DataFrame with at least 'asylum_location_code',
                           'year', 'month', and 'population' columns.

    Returns:
        pd.DataFrame: The DataFrame with added resource simulation columns.
    """
    df = df.copy() # Avoid modifying the original DataFrame

    # Ensure data is sorted for next-month calculations
    df = df.sort_values(by=['asylum_location_code', 'year', 'month']).reset_index(drop=True)

    resource_types = ['food', 'medicine', 'shelter', 'personnel']

    # Initialize resource columns
    for resource in resource_types:
        df[f'available_{resource}_perc'] = np.nan
        df[f'added_{resource}'] = 0 # Added resources start at 0 for each month

    # Simulate resources per asylum country
    for country_code in df['asylum_location_code'].unique():
        country_df_indices = df[df['asylum_location_code'] == country_code].index

        # Initialize resources for the first month of the country's data
        first_index = country_df_indices[0]
        for resource in resource_types:
            # Start with a random initial percentage (e.g., 50% to 100%)
            df.loc[first_index, f'available_{resource}_perc'] = np.random.uniform(50, 100)

        # Simulate for subsequent months for the country
        for i in range(len(country_df_indices) - 1):
            current_index = country_df_indices[i]
            next_index = country_df_indices[i+1]

            current_pop = df.loc[current_index, 'population']
            # Assuming required resources are proportional to population
            required_resources_base = current_pop / 1000 # Base requirement per 1000 people

            for resource in resource_types:
                current_available_perc = df.loc[current_index, f'available_{resource}_perc']

                # Calculate the absolute available resources based on current percentage and population
                # Let's assume 100% availability means resources for the current population * some factor
                # A simpler approach is to directly work with the percentage relative to required.
                # Required resources for the current month (scaled by a factor)
                required_current = required_resources_base * np.random.uniform(0.8, 1.2) # Add some variability

                # Available resources from the start of the month (based on previous month's availability)
                # The amount available at the start of the current month is what was available
                # at the end of the previous month (implied by the percentage).
                # We need to think about what the "100%" reference is.
                # Let's assume 100% available means resources sufficient for the *current* population requirement.
                # So, absolute available = (current_available_perc / 100) * required_current (conceptually)

                # However, the percentage itself is relative to the *required* amount for the current population.
                # So, resources available at the start of the month = (prev_available_perc / 100) * required_prev

                # Let's re-evaluate: available_resource_perc is (current_available_resource / current_required_resource) * 100
                # Amount available at the start of month M is the amount available at the end of month M-1.
                # Amount available at the end of month M-1 = (available_M-1_perc / 100) * required_M-1
                # Required amount for month M is required_M.
                # So, available_M_perc = (Amount available at start of month M) / required_M * 100

                # This implies we need the 'required' amount from the previous month.
                # Let's simplify the simulation: percentage changes based on population change and added resources.
                # The change in required resources is proportional to the change in population.
                # required_change = (df.loc[next_index, 'population'] - current_pop) / 1000

                # Available resources percentage for the next month depends on:
                # 1. The current available percentage
                # 2. The increase/decrease in population (affects the 'required' baseline)
                # 3. Resources consumed (assume fixed consumption per person per month)
                # 4. Added resources

                # Let's model the change in available percentage more directly:
                # Percentage change is influenced by population change and added resources.

                # Calculate required change (percentage points needed)
                # If population increases, the *same absolute amount* of resource represents a *lower percentage*
                # If population decreases, the *same absolute amount* of resource represents a *higher percentage*
                # Percentage change due to population:
                # New 100% reference = df.loc[next_index, 'population'] / 1000
                # Old 100% reference = current_pop / 1000
                # If current available amount was X, X = (current_available_perc / 100) * (current_pop / 1000)
                # New percentage = (X / (df.loc[next_index, 'population'] / 1000)) * 100
                # New percentage = ((current_available_perc / 100) * (current_pop / 1000)) / (df.loc[next_index, 'population'] / 1000) * 100
                # New percentage = current_available_perc * (current_pop / df.loc[next_index, 'population'])

                prev_available_perc = df.loc[current_index, f'available_{resource}_perc']
                prev_pop = df.loc[current_index, 'population']
                current_pop_next_month = df.loc[next_index, 'population']

                # Percentage adjustment due to population change (if population changes, the 'required' baseline changes)
                if prev_pop > 0: # Avoid division by zero
                    available_perc_after_pop_change = prev_available_perc * (prev_pop / current_pop_next_month)
                else: # If previous population was 0, assume 100% availability initially for the next month if pop > 0
                    available_perc_after_pop_change = 100 if current_pop_next_month > 0 else 0


                # Simulate consumption (decrease in percentage)
                # Assume a base consumption rate reduces percentage, plus some randomness
                consumption_decrease = np.random.uniform(5, 15) # e.g., 5-15 percentage points reduction due to consumption
                available_perc_after_consumption = available_perc_after_pop_change - consumption_decrease

                # Ensure percentage doesn't drop below 0
                available_perc_after_consumption = max(0, available_perc_after_consumption)

                # Simulate 'added resources' based on current percentage and population change
                # If available percentage is low OR population increased significantly, more resources might be added
                # Probability of adding resources increases if percentage is low or pop increased
                need_for_resources = (100 - available_perc_after_consumption) + (current_pop_next_month - prev_pop) / 10 # Simple heuristic
                probability_add = min(1.0, need_for_resources / 100) # Scale need to a probability

                added_amount_perc = 0
                if np.random.rand() < probability_add:
                     # If resources are added, simulate an amount that boosts the percentage
                    boost_amount_perc = np.random.uniform(10, 50) # Add 10-50 percentage points worth of resources
                    added_amount_perc = boost_amount_perc
                    df.loc[next_index, f'added_{resource}'] = boost_amount_perc # Record the *percentage boost* added

                # Final available percentage for the next month
                df.loc[next_index, f'available_{resource}_perc'] = available_perc_after_consumption + added_amount_perc

                # Ensure percentage stays within reasonable bounds (e.g., max 200% just in case of massive adds)
                df.loc[next_index, f'available_{resource}_perc'] = min(200, df.loc[next_index, f'available_{resource}_perc'])


    # Round percentages for cleaner data
    for resource in resource_types:
        df[f'available_{resource}_perc'] = df[f'available_{resource}_perc'].round(2)
        df[f'added_{resource}'] = df[f'added_{resource}'].round(2)


    return df

# Apply the function to your final DataFrame
df_final = simulate_resources(df_final)

print("\nDataFrame with simulated resources:")
print(df_final[['asylum_location_code', 'year', 'month', 'population',
                'available_food_perc', 'added_food',
                'available_medicine_perc', 'added_medicine',
                'available_shelter_perc', 'added_shelter',
                'available_personnel_perc', 'added_personnel']].head(50))

print("\nInfo of the final DataFrame after adding resource simulation columns:")
df_final.info()

# Optional: Save the updated DataFrame
output_path_final = '/content/drive/MyDrive/RefugeeAI Project Datasets/refugee_movement_monthly_with_conflicts_weather_resources.csv'
df_final.to_csv(output_path_final, index=False)
print(f"\nFinal DataFrame with simulated resources saved to '{output_path_final}'.")
```

```python
# prompt: Drop the duration column and year. I want you to encode the categorical columns in the resulting dataset and return two new dataframes. One in which origin_location_code is also included in the encoding while dropping asylum_location_code and another in which the case is reversed

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

def encode_categorical_columns(df, cols_to_encode, drop_cols=[]):
  """
    Encodes specified categorical columns using One-Hot Encoding and drops other columns.

    Args:
        df (pd.DataFrame): The input DataFrame.
        cols_to_encode (list): A list of column names to encode.
        drop_cols (list, optional): A list of additional column names to drop. Defaults to [].

    Returns:
        pd.DataFrame: The DataFrame with encoded columns and dropped columns.
  """
  df_encoded = df.drop(columns=drop_cols + ['duration', 'year']).copy()

  # Identify categorical columns to encode based on cols_to_encode
  categorical_cols = [col for col in cols_to_encode if col in df_encoded.columns]

  if not categorical_cols:
    print("Warning: No specified columns to encode found in the DataFrame.")
    return df_encoded

  # Initialize OneHotEncoder
  encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

  # Fit and transform the categorical columns
  encoded_data = encoder.fit_transform(df_encoded[categorical_cols])

  # Create a DataFrame with the encoded data and appropriate column names
  encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols), index=df_encoded.index)

  # Drop the original categorical columns
  df_encoded = df_encoded.drop(columns=categorical_cols)

  # Concatenate the original DataFrame (without original categorical columns) and the encoded DataFrame
  df_encoded = pd.concat([df_encoded, encoded_df], axis=1)

  return df_encoded

# Define categorical columns that are common to both scenarios
common_categorical_cols = [
    'origin_has_hrp', 'origin_in_gho', 'asylum_has_hrp', 'asylum_in_gho',
    'month']

# Scenario 1: Encode 'origin_location_code', drop 'asylum_location_code'
cols_to_encode_scenario1 = ['origin_location_code'] + common_categorical_cols
df_scenario1 = encode_categorical_columns(
    df_final.copy(), # Work on a copy to avoid modifying df_final for the second scenario
    cols_to_encode=cols_to_encode_scenario1,
    drop_cols=['asylum_location_code']
)

# Scenario 2: Encode 'asylum_location_code', drop 'origin_location_code'
cols_to_encode_scenario2 = ['asylum_location_code'] + common_categorical_cols
df_scenario2 = encode_categorical_columns(
    df_final.copy(), # Work on a fresh copy
    cols_to_encode=cols_to_encode_scenario2,
    drop_cols=['origin_location_code']
)

print("\n--- Scenario 1: Encoded origin_location_code, dropped asylum_location_code ---")
print(df_scenario1.head())
print("\nInfo for Scenario 1 DataFrame:")
df_scenario1.info()

print("\n--- Scenario 2: Encoded asylum_location_code, dropped origin_location_code ---")
print(df_scenario2.head())
print("\nInfo for Scenario 2 DataFrame:")
df_scenario2.info()
```

```python
def generate_capacities(size, min_val, max_val, peak_min, peak_max, concentration_factor=5):
    beta_values = np.random.beta(concentration_factor, concentration_factor, size)

    # Linearly scale beta values to the peak range [peak_min, peak_max]
    peak_range_values = peak_min + beta_values * (peak_max - peak_min)
    peak_center = (peak_min + peak_max) / 2
    peak_std = (peak_max - peak_min) / 3
    normal_values = np.random.normal(loc=peak_center, scale=peak_std, size=size)
    clipped_values = np.clip(normal_values, min_val, max_val)
    return clipped_values.astype(int)

min_capacity = 3000
max_capacity = 11000
peak_min_capacity = 7000
peak_max_capacity = 9000

# Create the 'capacity' column and fill with generated values
ngo_data['capacity'] = generate_capacities(
    len(ngo_data),
    min_capacity,
    max_capacity,
    peak_min_capacity,
    peak_max_capacity
)

print("\nNGO Data with 'capacity' column:")
display(ngo_data.head())

print("\nDescriptive statistics for 'capacity' column:")
display(ngo_data['capacity'].describe())
```

```python
import pandas as pd
import pulp

# Define the optimization problem
prob = pulp.LpProblem("Resource_Allocation", pulp.LpMinimize)
```

```python
# Variables: Amount of funding allocated to each region for each need type
regions = needs_data['region'].tolist()
need_types = ['food', 'health', 'education', 'skills', 'jobs', 'shelter', 'legal', 'mental support']
funding = pulp.LpVariable.dicts("Funding", [(r, n) for r in regions for n in need_types], lowBound=0, cat='Continuous')
```

```python
# NGO assignment: Binary variable indicating if NGO is assigned to a region and need type
ngos = ngo_data['ngo'].tolist()
assignment = pulp.LpVariable.dicts("Assignment", [(n, r, t) for n in ngos for r in regions for t in need_types], cat='Binary')
```

```python
# Objective: Minimize unmet needs (difference between needs and allocated resources)
# Assume $1 per person meets the need (simplified; adjust based on real costs)
prob += pulp.lpSum([
    needs_data.loc[needs_data['region'] == r, f"{n}"].iloc[0] - funding[(r, n)]
    for r in regions for n in need_types
]), "Minimize_Unmet_Needs"
```

```python
# Constraints
# 1. Funding constraints: Do not exceed available funding for each resource type
for n in need_types:
  display(funding_data.loc[funding_data['resource_type'] == n, 'available_funding'])
```

```python
# Constraints
# 1. Funding constraints: Do not exceed available funding for each resource type
for n in need_types:
    prob += (
        pulp.lpSum([funding[(r, n)] for r in regions]) <=
        funding_data.loc[funding_data['resource_type'] == n, 'available_funding'].iloc[0],
        f"Funding_Limit_{n}"
    )
```

```python
# 2. NGO capacity constraints: Respect NGO capacity in assigned regions
# Group by ngo and region to handle potential duplicates and ensure unique constraint names
grouped_ngo_data = ngo_data.groupby(['ngo', 'region']).agg({
    'capacity': 'max',  # Use max capacity if multiple entries for same NGO-region
    'focus_area': lambda x: ','.join(x.dropna().unique()) # Combine focus areas
}).reset_index()

# Ensure the grouped_ngo_data is unique on ngo and region before iterating
grouped_ngo_data = grouped_ngo_data.drop_duplicates(subset=['ngo', 'region']).reset_index(drop=True)


for index, row in grouped_ngo_data.iterrows():
    ngo = row['ngo']
    ngo_region = row['region']
    ngo_capacity = row['capacity']
    ngo_focus = row['focus_area'].split(',')

    if ngo_region in regions: # Ensure the region is one we are considering
        prob += (
            pulp.lpSum([funding[(ngo_region, t)] for t in need_types if t in ngo_focus]) <=
            ngo_capacity * pulp.lpSum([assignment[(ngo, ngo_region, t)] for t in need_types if t in ngo_focus]),
            f"Capacity_{index}_{ngo}_{ngo_region}" # Include index in constraint name for uniqueness
        )

# 3. NGO assignment constraints: Only assign NGOs to regions and needs matching their focus
# This part needs to consider the potential multiple focus areas from the grouped data
# Use the (potentially filtered and deduplicated) grouped_ngo_data for consistency
for index, row in grouped_ngo_data.iterrows():
    ngo = row['ngo']
    ngo_focus = row['focus_area'].split(',')
    ngo_region = row['region']
    for r in regions:
        for t in need_types:
            # Only allow assignment if the region matches the NGO's region AND the need type is in the NGO's focus
            if r != ngo_region or t not in ngo_focus:
                prob += assignment[(ngo, r, t)] == 0, f"No_Assignment_{index}_{ngo}_{r}_{t}" # Include index for uniqueness

# 4. Ensure needs are met only if an NGO is assigned
# This constraint remains the same as it is per region and need type, independent of specific NGOs
for r in regions:
    for t in need_types:
        prob += (
            funding[(r, t)] <=
            pulp.lpSum([assignment[(n, r, t)] * 1e6 for n in ngos]),  # Large number to allow sufficient funding
            f"Assignment_Required_{r}_{t}"
        )
```

```python
# Solve the problem
prob.solve()

# Output results
print("Allocation Plan:")
for r in regions:
    for n in need_types:
        if pulp.value(funding[(r, n)]) > 0:
            print(f"Region {r}, {n} needs: ${pulp.value(funding[(r, n)]):.2f} allocated")
for n in ngos:
    for r in regions:
        for t in need_types:
            if pulp.value(assignment[(n, r, t)]) == 1:
                print(f"NGO {n} assigned to Region {r} for {t} needs")
print(f"Total Unmet Needs: {pulp.value(prob.objective)} people")
```
